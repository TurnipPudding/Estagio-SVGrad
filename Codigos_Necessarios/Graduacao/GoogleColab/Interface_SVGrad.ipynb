### padroniza_dataframe

```python
def padroniza_dataframe(file_name, header_row, ano):
    # Leio o dataframe correto, com a linha do cabeçalho.
    df = pd.read_excel(file_name, header=header_row)

    # Primeiramente, para cada cabeçalho/coluna 'col' do dataframe.
    for col in range(len(df.columns)):
        # Verifico se tem um "\n" no texto do dataframe.
        if "\n" in df.columns[col]:
            # Se houver, substituo ele por um " ".
            df = df.rename(columns={df.columns[col] : df.columns[col].replace("\n", " ")})
        # Verifico se estou na coluna das salas - "Sala (a definir)".
        if "Sala" in df.columns[col]:
            # Se estiver, substituo o nome da coluna por apenas "Sala".
            df = df.rename(columns={df.columns[col] : "Sala"})

    for idx, row in df.iterrows():
        df.loc[idx, 'Deve ser alocada no ICMC?'] = str(df.loc[idx, 'Deve ser alocada no ICMC?']).replace(' ', '')
        if df.loc[idx, 'Deve ser alocada no ICMC?'] == 'x':
            df.loc[idx, 'Deve ser alocada no ICMC?'] = 'X'

    # Filtro o dataframe para fazer a edição apenas nas disciplinas que importam.
    df = df[df['Deve ser alocada no ICMC?'] == 'X']

    # Adiciona colunas de horário se não existirem
    if "Horário 3" not in df.columns:
        df.insert(df.columns.get_loc("Horário 2") + 1, "Horário 3", pd.NA)

    if "Horário 4" not in df.columns:
        df.insert(df.columns.get_loc("Horário 3") + 1, "Horário 4", pd.NA)

    if "observações" in df.columns:
        df = df.rename(columns={"observações" : "Observações"})

    if "Observações" not in df.columns:
        df.insert(df.columns.get_loc("Horário 4") + 1, "Observações", pd.NA)

    # Salvo o nome dos cabeçalhos do dataframe após as edições.
    headers = df.columns

    # Preencho os espaços vazios da coluna de uso de laboratórios com "Não"
    df['Utilizará laboratório? (sim ou não)'] = df['Utilizará laboratório? (sim ou não)'].fillna("Não")

    # Busco e listo por disciplinas cuja Turma não foi definida.
    turmas0 = df[df['Turma'].isna()].index.tolist()
    if len(turmas0) != 0:
        print(f"\nAVISO!\nA(s) linha(s) {[d + header_row + 2 for d in turmas0]} do arquivo {file_name} "
              f"possuem turmas não identificadas. Verifique qual a turma da disciplina.\n")
        return None

    # Adiciono mais duas colunas no dataframe
    df.insert(df.columns.get_loc(headers[-1]), "Vagas por disciplina", "")
    df.insert(df.columns.get_loc(headers[-1]), "Ano dos dados", ano)

    # Verificação e padronização dos horários
    for header in ['Horário 1', 'Horário 2', 'Horário 3', 'Horário 4']:
        for d in df.index:
            # Substitui traço anormal
            if "–" in str(df.loc[d, header]):
                df.loc[d, header] = df.loc[d, header].replace('–', '-')

            # Verifica formatação básica do horário
            if not pd.isna(df.loc[d, header]) \
                and "-" not in str(df.loc[d, header]) \
                    and not (df.loc[d, header] == ""):
                print(f"\nAVISO! Há um horário de aula não padronizado!\n"
                      f"Verifique a linha {d+header_row+2} da coluna {header} do arquivo {file_name}.\n"
                      f"Padrão correto: 'Segunda - 08:10/09:50'\n"
                      f"Provavelmente, falta um '-' entre o dia da semana e o horário da aula.\n")
                return None

            # Verifica múltiplos traços
            if not pd.isna(df.loc[d, header]) and len(str(df.loc[d, header]).split("-")) > 2:
                print(f"\nAVISO! Há um horário de aula não padronizado!\n"
                      f"Verifique a linha {d+header_row+2} da coluna {header} do arquivo {file_name}.\n"
                      f"Padrão correto: 'Segunda - 08:10/09:50'\n"
                      f"Provavelmente, há mais de um '-' na célula.\n")
                return None

    # Adiciona colunas para salas proibidas
    df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 1", pd.NA)
    df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 2", pd.NA)
    df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 3", pd.NA)
    df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 4", pd.NA)

    # Retorno o dataframe com todas as padronizações necessárias.
    return df
```

## Padrozina Dataframe

```python
import pandas as pd
from IPython.display import clear_output

def padroniza_dataframe(file_name, header_row, ano):
    """
    Função para padronizar dataframes no Google Colab
    Parâmetros:
    - file_name: caminho/nome do arquivo
    - header_row: linha do cabeçalho
    - ano: ano dos dados
    Retorna:
    - DataFrame padronizado ou None em caso de erro
    """
    try:
        # Leitura do arquivo
        # print(f"\n📊 Processando arquivo: {os.path.basename(file_name)}")
        df = pd.read_excel(file_name, header=header_row)

        # Padronização de cabeçalhos
        # print("🔄 Padronizando cabeçalhos...")
        for col in range(len(df.columns)):
            if "\n" in df.columns[col]:
                df = df.rename(columns={df.columns[col]: df.columns[col].replace("\n", " ")})
            if "Sala" in df.columns[col]:
                df = df.rename(columns={df.columns[col]: "Sala"})

        # Processamento da coluna 'Deve ser alocada no ICMC?'
        # print("🔍 Filtrando disciplinas para alocação...")
        for idx, row in df.iterrows():
            df.loc[idx, 'Deve ser alocada no ICMC?'] = str(df.loc[idx, 'Deve ser alocada no ICMC?']).replace(' ', '')
            if df.loc[idx, 'Deve ser alocada no ICMC?'] == 'x':
                df.loc[idx, 'Deve ser alocada no ICMC?'] = 'X'

        df = df[df['Deve ser alocada no ICMC?'] == 'X']

        # Adição de colunas padrão se não existirem
        # print("➕ Adicionando colunas padrão...")
        if "Horário 3" not in df.columns:
            df.insert(df.columns.get_loc("Horário 2") + 1, "Horário 3", pd.NA)

        if "Horário 4" not in df.columns:
            df.insert(df.columns.get_loc("Horário 3") + 1, "Horário 4", pd.NA)

        if "observações" in df.columns:
            df = df.rename(columns={"observações": "Observações"})

        if "Observações" not in df.columns:
            df.insert(df.columns.get_loc("Horário 4") + 1, "Observações", pd.NA)

        headers = df.columns

        # Preenchimento de valores padrão
        df['Utilizará laboratório? (sim ou não)'] = df['Utilizará laboratório? (sim ou não)'].fillna("Não")

        # Verificação de turmas não definidas
        turmas0 = df[df['Turma'].isna()].index.tolist()
        if len(turmas0) != 0:
            print(f"\n⚠️ AVISO: Turmas não identificadas no arquivo {os.path.basename(file_name)}")
            print(f"Linhas com problema: {[d + header_row + 2 for d in turmas0]}")
            print("Por favor, verifique qual a turma da disciplina.")
            return None

        # Adição de colunas adicionais
        df.insert(df.columns.get_loc(headers[-1]), "Vagas por disciplina", "")
        df.insert(df.columns.get_loc(headers[-1]), "Ano dos dados", ano)

        # Padronização de horários
        # print("⏱️ Verificando formatação de horários...")
        for header in ['Horário 1', 'Horário 2', 'Horário 3', 'Horário 4']:
            for d in df.index:
                # Substituição de traços anormais
                if "–" in str(df.loc[d, header]):
                    df.loc[d, header] = df.loc[d, header].replace('–', '-')

                # Validação do formato dos horários
                if not pd.isna(df.loc[d, header]) and "-" not in str(df.loc[d, header]) and str(df.loc[d, header]).strip():
                    print(f"\n⚠️ AVISO: Horário não padronizado encontrado!")
                    print(f"Arquivo: {os.path.basename(file_name)}")
                    print(f"Linha: {d + header_row + 2}")
                    print(f"Coluna: {header}")
                    print(f"Conteúdo: '{df.loc[d, header]}'")
                    print("Padrão esperado: 'Segunda - 08:10/09:50'")
                    print("Provavelmente falta um '-' entre o dia da semana e o horário da aula.")
                    return None

                if not pd.isna(df.loc[d, header]) and len(str(df.loc[d, header]).split("-")) > 2:
                    print(f"\n⚠️ AVISO: Formato de horário inválido!")
                    print(f"Arquivo: {os.path.basename(file_name)}")
                    print(f"Linha: {d + header_row + 2}")
                    print(f"Coluna: {header}")
                    print(f"Conteúdo: '{df.loc[d, header]}'")
                    print("Padrão esperado: 'Segunda - 08:10/09:50'")
                    print("Provavelmente há mais de um '-' na célula.")
                    return None

        # Adiciono 4 colunas para anotar possíveis salas onde as aulas ficam PROIBIDAS de serem alocadas.
        # Cada coluna refere-se a uma coluna de horários, fazendo com que as salas listadas em uma coluna afetem apenas a aula de mesma coluna.
        # Ex: Adicionar a sala 3-009 na segunda coluna de proibição faz com que a aula do Horário 2 de uma disciplina não possa ser na 3-009.
        df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 1", pd.NA)
        df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 2", pd.NA)
        df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 3", pd.NA)
        df.insert(df.columns.get_loc(headers[-1]), "Proibir Horário 4", pd.NA)

        print("✅ Padronização concluída com sucesso!")
        return df

    except Exception as e:
        print(f"\n❌ Erro durante a padronização do dataframe:")
        print(f"Arquivo: {os.path.basename(file_name)}")
        print(f"Erro: {str(e)}")
        return None
```

## ler_df

```python
# Função que lê um dado arquivo.
def ler_df(caminho_arquivo):
    """
    Função para ler arquivos de dados (Excel ou CSV) no Google Colab
    Retorna um DataFrame pandas ou None em caso de erro
    """
    try:
        if caminho_arquivo.endswith(('.xlsx', '.xls')):
            df = pd.read_excel(caminho_arquivo)
            return df
        elif caminho_arquivo.endswith('.csv'):
            df = pd.read_csv(caminho_arquivo, on_bad_lines='skip', sep=';', encoding='latin-1')
            return df
        else:
            print("❌ Erro: Formato de arquivo não suportado!")
            print("Por favor, use arquivos .xlsx, .xls ou .csv")
            return None

    except Exception as e:
        print(f"❌ Erro ao carregar o arquivo {caminho_arquivo}:\n\n{e}")
        return None
```

concat_df

```python


def concat_df(SME, SMA, SCC, SSC, salas, nome_arquivo, ano, jupiter, outros):
    """
    Função para concatenar planilhas em um único arquivo no Google Colab
    """
    files = []

    if not jupiter:
        # Lê a planilha de salas
        print("\n📂 Lendo dados das salas...")
        df_salas = ler_df(salas)
        if df_salas is None:
            print("❌ Erro na leitura do arquivo de salas. Operação cancelada.")
            return

        name_files = [SME, SMA, SCC, SSC, outros]
        sheets = ["SME", "SMA", "SCC", "SSC", "Outros"]
    else:
        name_files = [SME, SMA, SCC, SSC]
        name_files.extend(outros)
        sheets = []

    # Processa cada arquivo
    for i, name in enumerate(name_files):
        # print(f"\n📂 Processando arquivo {i+1}/{len(name_files)}...")
        df = ler_df(name)
        if df is None:
            print(f"❌ Erro no arquivo {name}. Operação cancelada.")
            return

        if not jupiter:
            header_name = 'Disciplina (código)'
            header_found = False

            for i, valor in enumerate(df.loc[:,df.columns[0]]):
                if valor == header_name:
                    header_row = i+1
                    df_padronizado = padroniza_dataframe(name, header_row, ano)

                    if df_padronizado.empty:
                        print(f"❌ O arquivo '{os.path.basename(name)}' não pode ser lido ou está vazio.")
                        print("Verifique se as disciplinas a serem alocadas estão marcadas.")
                        return

                    files.append(df_padronizado)
                    header_found = True
                    break

            if not header_found:
                print(f"❌ A coluna '{header_name}' não foi encontrada no arquivo {os.path.basename(name)}")
                print("Verifique o cabeçalho do arquivo de entrada.")
                return
        else:
            try:
                new_name = str(df.loc[0, 'Disciplina'])[:3]
                sheets.append(new_name)
                files.append(df)
            except KeyError:
                print(f"❌ A coluna 'Disciplina' não foi encontrada no arquivo {os.path.basename(name)}")
                print("Verifique o cabeçalho do arquivo de entrada.")
                return
            except Exception as e:
                print(f"❌ Ocorreu um erro inesperado:\n{e}")
                return

    # Garante a extensão .xlsx
    if not nome_arquivo.endswith(".xlsx"):
        nome_arquivo += ".xlsx"

    # Caminho de saída no Google Drive

    output_path = os.path.join('/content/drive/MyDrive/Estagio-SVGrad/Saídas da Interface/Planilhas de Dados', nome_arquivo)

    try:
        # print("\n🔄 Salvando arquivo concatenado...")
        with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
            if not jupiter:
                df_salas.to_excel(writer, sheet_name="Salas", index=False)

            for sh in range(len(files)):
                files[sh].to_excel(writer, sheet_name=sheets[sh], index=False)

        print(f"\n✅ Arquivo {nome_arquivo} criado com sucesso!")
        print(f"Local: {output_path}")

    except PermissionError:
        print(f"❌ Não foi possível salvar o arquivo {nome_arquivo}")
        print("Verifique se ele está aberto em outro programa e tente novamente.")
    except Exception as e:
        print(f"❌ Ocorreu um erro inesperado:\n{e}")
```

## planilha_dep

```python
from google.colab import files
from IPython.display import clear_output
import os

def planilha_dep(jupiter):
    """
    Função para construção de planilhas de dados usando upload interativo no Colab
    Parâmetro:
    - jupiter: booleano que indica se é para dados do JúpiterWeb (True) ou das Aulas (False)
    """

    clear_output(wait=True)
    print(f"=== CONSTRUIR {'BASE DE DADOS DO JÚPITERWEB' if jupiter else 'BASE DE DADOS DAS AULAS'} ===")

    # Dicionário para armazenar os arquivos carregados
    arquivos_carregados = {}

    # Função auxiliar para upload com mensagem específica
    def fazer_upload(mensagem):
        print(f"\n{mensagem}")
        print("Por favor, faça o upload do arquivo:")
        uploaded = files.upload()
        if not uploaded:
            print("Nenhum arquivo foi selecionado!")
            return None
        # Retorna o primeiro arquivo carregado
        return next(iter(uploaded))

    # Solicitar arquivos obrigatórios
    arquivos_carregados['sme'] = fazer_upload("1. Planilha do SME (arquivo Excel)")
    arquivos_carregados['sma'] = fazer_upload("2. Planilha do SMA (arquivo Excel)")
    arquivos_carregados['scc'] = fazer_upload("3. Planilha do SCC (arquivo Excel)")
    arquivos_carregados['ssc'] = fazer_upload("4. Planilha do SSC (arquivo Excel)")

    if not jupiter:
        # Arquivos específicos para dados das aulas
        arquivos_carregados['outros'] = fazer_upload("5. Planilha dos Outros Institutos (arquivo Excel)")
        arquivos_carregados['salas'] = fazer_upload("6. Planilha dos dados das salas (arquivo Excel)")
        ano_dados = input("\n7. Digite o ano dos dados: ").strip()
        # ano_dados = input_widget("7. Digite o ano dos dados:")
    else:
        # Lógica para múltiplos arquivos no modo Júpiter
        lista_outros = []
        print("\n5. Arquivos dos Outros Institutos (faça upload de cada arquivo separadamente)")
        while True:
            print(f"\nArquivo {len(lista_outros)+1} (deixe vazio e pressione Enter para parar):")
            uploaded = files.upload()
            if not uploaded:
                break
            lista_outros.append(next(iter(uploaded)))

    nome_arquivo = input("\nDigite o nome para a base de dados (sem extensão): ").strip()
    # nome_arquivo = input_widget("Digite o nome para a base de dados:")

    # Validação dos inputs
    def validar_arquivo(arquivo, nome):
        if arquivo is None:
            print(f"Erro: Por favor, forneça o {nome}")
            return False
        return True

    # Verifica arquivos obrigatórios
    if not all([
        validar_arquivo(arquivos_carregados['sme'], "arquivo do SME"),
        validar_arquivo(arquivos_carregados['sma'], "arquivo do SMA"),
        validar_arquivo(arquivos_carregados['scc'], "arquivo do SCC"),
        validar_arquivo(arquivos_carregados['ssc'], "arquivo do SSC")
    ]):
        input("\nPressione Enter para tentar novamente...")
        return planilha_dep(jupiter)

    if not jupiter:
        if not all([
            validar_arquivo(arquivos_carregados['outros'], "arquivo dos Outros Institutos"),
            validar_arquivo(arquivos_carregados['salas'], "arquivo dos dados das salas"),
            ano_dados
        ]):
            input("\nPressione Enter para tentar novamente...")
            return planilha_dep(jupiter)
    else:
        if not lista_outros:
            print("Erro: Por favor, adicione pelo menos um arquivo dos Outros Institutos")
            input("\nPressione Enter para tentar novamente...")
            return planilha_dep(jupiter)

    if not nome_arquivo:
        print("Erro: Por favor, forneça um nome para a base de dados")
        input("\nPressione Enter para tentar novamente...")
        return planilha_dep(jupiter)

    # Chamar a função de concatenação
    if not jupiter:
        concat_df(
            arquivos_carregados['sme'], arquivos_carregados['sma'], arquivos_carregados['scc'],
            arquivos_carregados['ssc'], arquivos_carregados['salas'], nome_arquivo,
            ano_dados, jupiter, arquivos_carregados['outros']
        )
    else:
        concat_df(
            arquivos_carregados['sme'], arquivos_carregados['sma'], arquivos_carregados['scc'],
            arquivos_carregados['ssc'], None, nome_arquivo,
            None, jupiter, lista_outros
        )

    print("\n✅ Base de dados criada com sucesso!")
    input("\nPressione Enter para voltar ao menu principal...")

# planilha_dep(False)
```

```python
import pandas as pd
import os
import subprocess
import sys
from google.colab import files
from IPython.display import clear_output

def base_dados(pior_caso):
    """
    Função para construção de base de dados no Google Colab
    Parâmetro:
    - pior_caso: booleano que indica se é para criar base de pior caso (True) ou base normal (False)
    """

    clear_output(wait=True)
    print(f"=== CONSTRUIR {'BASE DE DADOS DE PIOR CASO' if pior_caso else 'BASE DE DADOS DO MODELO'} ===")

    # Função auxiliar para upload com mensagem específica
    def fazer_upload(mensagem):
        print(f"\n{mensagem}")
        print("Por favor, faça o upload do arquivo:")
        uploaded = files.upload()
        if not uploaded:
            print("❌ Nenhum arquivo foi selecionado!")
            return None
        # Retorna o primeiro arquivo carregado
        return next(iter(uploaded))

    # Dicionário para armazenar os arquivos carregados
    arquivos = {}

    if not pior_caso:
        # Caso normal: pede 4 arquivos
        print("\n📂 Selecione os arquivos necessários:")
        arquivos['base'] = fazer_upload("1. Planilha dos dados das aulas")
        arquivos['jptr'] = fazer_upload("2. Planilha dos dados do Júpiter")
        arquivos['ing'] = fazer_upload("3. Planilha dos dados dos ingressantes")
        arquivos['esp'] = fazer_upload("4. Planilha com os inscritos das disciplinas com espelho")
    else:
        # Caso pior caso: pede 2 arquivos
        print("\n📂 Selecione as bases para comparação:")
        arquivos['base1'] = fazer_upload("1. Base de dados mais recente")
        arquivos['base2'] = fazer_upload("2. Base de dados mais antiga")

    # Solicitar nome do arquivo de saída
    nome_arquivo = input("\nDigite o nome para a nova base de dados (sem extensão): ").strip()
    # nome_arquivo = input_widget("Digite o nome para a nova base de dados:")

    # Validação dos inputs
    def validar_arquivo(arquivo, nome):
        if arquivo is None:
            print(f"❌ Por favor, forneça o {nome}")
            return False
        return True

    if not pior_caso:
        if not all([
            validar_arquivo(arquivos['base'], "arquivo dos dados das aulas"),
            validar_arquivo(arquivos['jptr'], "arquivo dos dados do Júpiter"),
            validar_arquivo(arquivos['ing'], "arquivo dos dados dos ingressantes"),
            validar_arquivo(arquivos['esp'], "arquivo dos inscritos em disciplinas com espelho"),
            nome_arquivo
        ]):
            input("\nPressione Enter para tentar novamente...")
            # input_widget("Pressione Enter para tentar novamente...")
            return base_dados(pior_caso)
    else:
        if not all([
            validar_arquivo(arquivos['base1'], "base de dados mais recente"),
            validar_arquivo(arquivos['base2'], "base de dados mais antiga"),
            nome_arquivo
        ]):
            input("\nPressione Enter para tentar novamente...")
            return base_dados(pior_caso)

    # Garante a extensão .xlsx
    if not nome_arquivo.endswith(".xlsx"):
        nome_arquivo += ".xlsx"

    # Caminho de saída no Google Drive
    output_path = os.path.join('/content/drive/MyDrive/Estagio-SVGrad/Saídas da Interface/Planilhas de Dados', nome_arquivo)
    script_path = os.path.join('/content/Estagio-SVGrad/Codigos_Necessarios/Graduacao/GoogleColab/')

    try:
        if not pior_caso:
            print("\n🔄 Processando dados para criar base do modelo...")
            # Executa o script de processamento
            result = subprocess.run(
                [sys.executable, script_path + "jupiter sheet maker colab.py",
                 arquivos['base'], arquivos['jptr'], arquivos['ing'], arquivos['esp'], output_path],
                check=True,
                capture_output=True,
                text=True
            )

            if result.returncode == 0:
                print(f"\n✅ Base de dados criada com sucesso em:\n{output_path}")
            else:
                print(f"\n❌ Erro ao executar o script:\n{result.stderr}")
        else:
            print("\n🔄 Criando base de pior caso...")
            # Lê as bases de dados
            df1 = pd.read_excel(arquivos['base1'], sheet_name=["Salas", "SME", "SMA", "SCC", "SSC", "Outros"])
            df2 = pd.read_excel(arquivos['base2'], sheet_name=["Salas", "SME", "SMA", "SCC", "SSC", "Outros"])

            # Chama a função para criar base de pior caso
            df_pior_caso = base_pior_caso(df1, df2, ["Salas", "SME", "SMA", "SCC", "SSC", "Outros"])

            # Salva o resultado
            with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
                for sh, df_sh in df_pior_caso.items():
                    df_sh.to_excel(writer, sheet_name=sh, index=False)

            print(f"\n✅ Base de pior caso criada com sucesso em:\n{output_path}")

    except PermissionError:
        print(f"\n❌ Não foi possível salvar o arquivo {nome_arquivo}")
        print("Verifique se ele está aberto em outro programa e tente novamente.")
    except subprocess.CalledProcessError as e:
        print(f"\n❌ Erro ao executar o script:")
        print(e.stderr if e.stderr else "Erro desconhecido")
    except Exception as e:
        print(f"\n❌ Ocorreu um erro inesperado:")
        print(str(e))
    finally:
        input("\nPressione Enter para voltar ao menu principal...")
        # Input = input_widget("Pressione Enter para voltar ao menu principal...")
```

# Botão 4 - Pior Caso

```python
def base_pior_caso(df1, df2, sheets):
    """
    Função para criar base de dados de pior caso comparando duas bases
    Parâmetros:
    - df1: DataFrame da base mais recente (dicionário de DataFrames por sheet)
    - df2: DataFrame da base mais antiga (dicionário de DataFrames por sheet)
    - sheets: Lista de nomes das planilhas a serem processadas
    Retorna:
    - Dicionário de DataFrames com os valores de pior caso ou None em caso de erro
    """

    print("\n🔍 Analisando bases para determinar pior caso...")

    try:
        # Processa cada planilha (exceto a primeira que é 'Salas')
        for sh in sheets[1:]:
            # print(f"📊 Processando planilha: {sh}...")

            # Obtém as planilhas correspondentes de ambas as bases
            df1_sh = df1[sh]
            df2_sh = df2[sh]

            # Converte colunas de disciplinas para listas
            l1 = df1_sh['Disciplina (código)'].tolist()
            l2 = df2_sh['Disciplina (código)'].tolist()

            # Para cada disciplina na base mais recente
            for d in l1:
                if d in l2:
                    # Compara número de vagas
                    vagas_df1 = df1_sh.loc[l1.index(d), 'Vagas por disciplina']
                    vagas_df2 = df2_sh.loc[l2.index(d), 'Vagas por disciplina']

                    if vagas_df1 < vagas_df2:
                        # Atualiza para o pior caso
                        df1_sh.loc[l1.index(d), 'Vagas por disciplina'] = vagas_df2
                        df1_sh.loc[l1.index(d), 'Ano dos dados'] = df2_sh.loc[l2.index(d), 'Ano dos dados']
                        # print(f"  ⚠️ Atualizada disciplina {d} para pior caso: {vagas_df2} vagas")

        # print("✅ Análise de pior caso concluída com sucesso!")
        return df1

    except KeyError as e:
        coluna_faltando = str(e).strip("'")
        print(f"\n❌ Erro: Coluna '{coluna_faltando}' não encontrada")
        print("Verifique se os arquivos possuem a estrutura esperada:")
        print("- Coluna 'Disciplina (código)' para identificar as disciplinas")
        print("- Coluna 'Vagas por disciplina' com os números de inscritos")
        print("- Coluna 'Ano dos dados' com o ano de referência")
        return None

    except Exception as e:
        print(f"\n❌ Ocorreu um erro inesperado durante a análise:")
        print(str(e))
        return None
```

### roda_script

```python
def roda_script(script, nome, peso_x, peso_y, peso_v, peso_z, alpha, pref):
    """
    Executa um script Python e exibe sua saída no Google Colab

    Parâmetros:
    - script: Nome do script a ser executado
    - nome: Nome do arquivo de entrada principal
    - peso_x, peso_y, peso_v, peso_z, alpha, pref: Parâmetros adicionais para o script
    """
    folder_path = '/content/drive/MyDrive/Estagio-SVGrad/'
    clear_output(wait=True)
    print(f"=== EXECUTANDO SCRIPT: {script} ===")
    print(f"Arquivo de entrada: {nome}")
    print("Parâmetros:")
    print(f"- peso_x: {peso_x}")
    print(f"- peso_y: {peso_y}")
    print(f"- peso_v: {peso_v}")
    print(f"- peso_z: {peso_z}")
    print(f"- alpha: {alpha}")
    print(f"- pref: {pref}")
    print(f"Caminho da pasta: {}")
    print("\n" + "="*40 + "\n")

    try:
        # Configura o processo
        process = subprocess.Popen(
            [sys.executable, script, nome, str(peso_x), str(peso_y),
             str(peso_v), str(peso_z), str(alpha), str(pref), str(folder_path)],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True,
            bufsize=1,  # Buffer de linha
            universal_newlines=True
        )

        # Exibe a saída em tempo real
        print("🚀 Iniciando execução...\n")
        print("-"*40)

        while True:
            output = process.stdout.readline()
            if output == '' and process.poll() is not None:
                break
            if output:
                print(output.strip())

        print("-"*40)

        # Verifica erros
        stderr = process.communicate()[1]
        return_code = process.returncode

        if return_code != 0:
            print("\n❌ ERRO DURANTE EXECUÇÃO:")
            print(stderr)

            if return_code == 1:
                print("\n💡 Dica: Verifique os arquivos de entrada e parâmetros")
            elif return_code == 2:
                print(f"\n💡 Dica: Verifique se o arquivo {nome} está acessível")
        # else:
            # print("\n✅ Script executado com sucesso!")

    except subprocess.CalledProcessError as e:
        print(f"\n❌ Erro ao executar o script (código {e.returncode}):")
        print(e.stderr if e.stderr else "Erro desconhecido")
    except Exception as e:
        print(f"\n❌ Erro inesperado: {str(e)}")
    finally:
        input("\nPressione Enter para continuar...")
```

### criar_visualizacao_de_vazias

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from datetime import datetime, timedelta
import os

def criar_visualizacao_de_vazias(file_path, file_path_salas, caminho_arquivo, caminho_arquivo1):
    """
    Função para criar visualização de espaços livres no Google Colab
    Gera dois arquivos:
    1. Planilha com visualização gráfica dos horários livres
    2. Planilha com lista detalhada dos horários livres
    """

    try:
        print("\n📊 Processando arquivos...")

        # Carrega a planilha de solução
        wb = load_workbook(file_path)
        ws = wb.active

        # Define intervalo de horários (7:00 às 23:30 em intervalos de 30 minutos)
        start_time = datetime.strptime("07:00", "%H:%M")
        end_time = datetime.strptime("23:30", "%H:%M")
        horarios = []
        current_time = start_time
        while current_time <= end_time:
            horarios.append(current_time.strftime("%H:%M"))
            current_time += timedelta(minutes=30)

        sala_colunas = len(horarios)
        dias_semana = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado"]

        # Configurações iniciais para processamento
        start_row = 1
        start_column = 2
        end_row = 1
        end_column = 1 + sala_colunas
        start = start_row + 2
        space_between = 3

        # Estilo para células livres (amarelo)
        yellow_fill = PatternFill(start_color="00E3DE00", end_color="00E3DE00", fill_type="solid")

        # Calcula número de salas utilizadas
        n_salas = int((ws.max_row+1) / (2 + len(dias_semana) + space_between))

        # Carrega dados das salas
        salas = pd.read_excel(file_path_salas)
        salas1 = []
        df_vazio = pd.DataFrame(columns=["Sala", "Dia da semana", "Horário vago"])

        # Função auxiliar para verificar cor da célula
        def get_cell_color(cell):
            fill = cell.fill
            if fill and fill.start_color.index != "00E3DE00":  # Ignorar células amarelas
                return fill.start_color.index
            return None

        print("🔍 Analisando horários livres por sala...")

        # Processa cada sala utilizada
        for i in range(n_salas+1):
            sala = ws.cell(row=start_row, column=start_column).value
            salas1.append(sala)

            # Processa cada dia da semana
            for row in ws.iter_rows(min_row=start, max_row=start+5):
                start_col = None

                for col_idx, cell in enumerate(row, start=1):
                    cell_color = get_cell_color(cell)

                    # Célula vazia (potencial horário livre)
                    if cell_color == "00000000":
                        if type(cell) != openpyxl.cell.cell.MergedCell:
                            if start_col is None and cell.value not in dias_semana:
                                start_col = cell.column - 2

                    # Célula ocupada
                    else:
                        if type(cell) != openpyxl.cell.cell.MergedCell:
                            cell.value = ""
                            cell.fill = yellow_fill

                        # Registra intervalo livre encontrado
                        if start_col is not None and col_idx - start_col >= 1:
                            df_vazio.loc[len(df_vazio)] = [
                                f"{sala}",
                                f"{ws.cell(row=cell.row, column=1).value}",
                                f"{horarios[start_col]} - {horarios[col_idx-2]}"
                            ]
                        start_col = None

                # Registra último intervalo livre do dia
                if start_col is not None and start_col < len(row):
                    df_vazio.loc[len(df_vazio)] = [
                        f"{sala}",
                        f"{ws.cell(row=cell.row, column=1).value}",
                        f"{horarios[start_col]} - {horarios[col_idx-2]}"
                    ]

            # Atualiza para próxima sala
            start_row = start_row + 2 + len(dias_semana) + space_between
            end_row = start_row
            start = start_row + 2

        print("🔍 Verificando salas não utilizadas...")

        # Processa salas não utilizadas
        for sala in salas['Sala'].tolist():
            if sala not in salas1:
                # Ignora salas individuais quando o conjunto foi usado
                if (sala == '6-303' and '6-303/6-304' in salas1) or \
                   (sala == '6-304' and '6-303/6-304' in salas1) or \
                   (sala == '6-305' and '6-305/6-306' in salas1) or \
                   (sala == '6-306' and '6-305/6-306' in salas1):
                    continue

                # Adiciona sala não utilizada (livre o dia todo)
                for dia in dias_semana:
                    df_vazio.loc[len(df_vazio)] = [
                        f"{sala}",
                        dia,
                        "07:00 - 23:30"
                    ]

        # Garante extensão .xlsx
        if not caminho_arquivo.endswith(".xlsx"):
            caminho_arquivo += ".xlsx"
        if not caminho_arquivo1.endswith(".xlsx"):
            caminho_arquivo1 += ".xlsx"

        # Caminhos completos no Google Drive
        output_path1 = os.path.join('/content/drive/MyDrive/Estagio-SVGrad/Saídas da Interface/Planilhas de Dados', caminho_arquivo)
        output_path2 = os.path.join('/content/drive/MyDrive/Estagio-SVGrad/Saídas da Interface/Planilhas de Dados', caminho_arquivo1)

        print("💾 Salvando arquivos...")

        # Salva arquivos
        wb.save(output_path1)
        df_vazio.to_excel(output_path2, sheet_name="Resultados", index=False)

        print("\n✅ Arquivos gerados com sucesso!")
        print(f"- Visualização gráfica: {output_path1}")
        print(f"- Lista detalhada: {output_path2}")

    except PermissionError:
        print("\n❌ Erro: Não foi possível salvar os arquivos")
        print("Verifique se os arquivos estão abertos em outro programa")
    except Exception as e:
        print(f"\n❌ Ocorreu um erro inesperado:")
        print(str(e))
```

### escolhas_preenchimento

```python
from google.colab import files
from IPython.display import clear_output
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

def escolhas_preenchimento(elenco1, file_path_sol1, file_path_base1):
    """
    Função para selecionar aulas a serem fixadas e preencher planilhas
    Parâmetros:
    - elenco1: Lista de caminhos para planilhas dos departamentos
    - file_path_sol1: Caminho da planilha com dados da solução
    - file_path_base1: Caminho da base de dados
    """

    clear_output(wait=True)
    print("=== SELECIONAR AULAS PARA FIXAR ===")

    # Carregar dados da solução
    try:
        df = pd.read_excel(file_path_sol1)
        df = df.sort_values(by=['Disciplina'], ignore_index=True)
    except Exception as e:
        print(f"❌ Erro ao carregar arquivo de solução: {str(e)}")
        input("\nPressione Enter para voltar...")
        return

    # Criar interface interativa
    print("\n🔍 Selecione as aulas que deseja fixar (manter na mesma sala):")

    # Widgets para seleção
    checkboxes = []
    output = widgets.Output()

    with output:
        # Cabeçalho
        header = widgets.HBox([
            widgets.Label(value="Disciplina", layout=widgets.Layout(width='300px')),
            widgets.Label(value="Horário", layout=widgets.Layout(width='150px')),
            widgets.Label(value="Sala", layout=widgets.Layout(width='100px')),
            widgets.Label(value="Inscritos", layout=widgets.Layout(width='80px')),
            widgets.Label(value="Fixar", layout=widgets.Layout(width='80px'))
        ])
        display(header)

        # Linhas com dados e checkboxes
        for d in df.index:
            hbox = widgets.HBox([
                widgets.Label(value=str(df.loc[d, 'Disciplina']), layout=widgets.Layout(width='300px')),
                widgets.Label(value=str(df.loc[d, 'Horário']), layout=widgets.Layout(width='150px')),
                widgets.Label(value=str(df.loc[d, 'Sala']), layout=widgets.Layout(width='100px')),
                widgets.Label(value=str(df.loc[d, 'Inscritos']), layout=widgets.Layout(width='80px')),
                widgets.Checkbox(value=False, layout=widgets.Layout(width='80px'))
            ])
            checkboxes.append(hbox)
            display(hbox)

    # Botão para confirmar
    button = widgets.Button(description="Preencher planilhas com salas fixadas")

    def on_button_click(b):
        clear_output(wait=True)
        print("🔄 Processando seleção...")

        # Criar novo DataFrame com aulas selecionadas
        new_df = pd.DataFrame(columns=df.columns)
        for i, hbox in enumerate(checkboxes):
            if hbox.children[4].value:  # Verifica se o checkbox está marcado
                new_df.loc[len(new_df)] = df.loc[i]

        if new_df.empty:
            print("⚠️ Nenhuma aula foi selecionada para fixar!")
            input("\nPressione Enter para tentar novamente...")
            return escolhas_preenchimento(elenco1, file_path_sol1, file_path_base1)

        try:
            # Salvar arquivo temporário com aulas fixadas
            temp_path = file_path_sol1.replace('.xlsx',' com Fixadas.xlsx')
            new_df.to_excel(temp_path, sheet_name="Resultados", index=False)

            # Chamar função de preenchimento
            preenchimento(elenco1, temp_path, file_path_base1, False)

            print("\n✅ Planilhas preenchidas com sucesso!")
            print(f"- Aulas fixadas: {len(new_df)}")

        except PermissionError:
            print("❌ Erro: Não foi possível salvar o arquivo")
            print("Verifique se o arquivo está aberto em outro programa")
        except Exception as e:
            print(f"❌ Ocorreu um erro inesperado:")
            print(str(e))

        input("\nPressione Enter para voltar ao menu principal...")

    button.on_click(on_button_click)

    # Exibir interface
    display(output, button)
```

### preenchimento

```python
def preenchimento(lista_elenco, file_path_sol, file_path_base, preencher_elenco):
    # Abro a solução dada pelo modelo como um dataframe.
    solucao = pd.read_excel(file_path_sol)

    if preencher_elenco:
        # Para cada elenco na lista de elencos:
        for file_path_elenco in lista_elenco:
            # Leio e salvo o arquivo em uma variável.
            elenco = pd.read_excel(file_path_elenco)

            # Defino uma variável com o nome de um cabeçalho para ser encontrado, caso o cabeçalho não seja a primeira linha da planilha.
            header_name = 'Disciplina (código)'

            # Para cada linha e célula da primeira coluna do dataframe:
            for i, valor in enumerate(elenco.loc[:,elenco.columns[0]]):
                # Se o valor da célula for o nome do cabeçalho que estou procurando:
                if valor == header_name:
                    # Salvo o número da linha do cabeçalho.
                    header_row = i+1
                    # E interrompo o loop.
                    break

            # Leio o arquivo da forma correta.
            elenco = pd.read_excel(file_path_elenco, header=header_row)

            # Para cada coluna da planilha:
            for col in range(len(elenco.columns)):
                # Verifico se tem um "\n" no texto do dataframe:
                if "\n" in elenco.columns[col]:
                    # Se houver, substituo ele por um " ".
                    elenco = elenco.rename(columns={elenco.columns[col] : elenco.columns[col].replace("\n", " ")})
                # Verifico se estou na coluna das salas:
                if "Sala" in elenco.columns[col]:
                    # Se estiver, substituo o nome da coluna por apenas "Sala"
                    elenco = elenco.rename(columns={elenco.columns[col] : "Sala"})

            # Abrir o arquivo como workbook
            wb = load_workbook(file_path_elenco)
            ws = wb.active

            # Para cada linha do dataframe:
            for d in range(len(elenco)):
                # Garanto que o código de uma disciplina não possui espaços.
                elenco.loc[d, 'Disciplina (código)'] = str(elenco.loc[d, 'Disciplina (código)']).replace(' ', '')

                # Adiciono um traço '-' e o número da turma ao nome da disciplina
                disciplina = str(elenco.loc[d, 'Disciplina (código)']) + '-' + str(elenco.loc[d, 'Turma'])

                # Verifico se a disciplina em questão foi incluída na solução:
                if disciplina in solucao['Disciplina'].tolist():
                    row = d+2
                    coluna = elenco.columns.get_loc('Sala')+1
                    ws.cell(row=header_row+row, column=coluna).value = None

                    solucao_filtrada = solucao[solucao['Disciplina'] == disciplina]

                    for i in range(len(solucao_filtrada)):
                        for a in solucao_filtrada.index:
                            if elenco.loc[d, 'Horário ' + str(i+1)] == solucao_filtrada.loc[a, 'Horário']:
                                if ws.cell(row=header_row+row, column=coluna).value:
                                    novo_valor = str(ws.cell(row=header_row+row, column=coluna).value) + ', ' + str(solucao_filtrada.loc[a, 'Sala'])
                                    ws.cell(row=header_row+row, column=coluna).value = novo_valor
                                else:
                                    ws.cell(row=header_row+row, column=coluna).value = str(solucao_filtrada.loc[a, 'Sala'])

            try:
                output_path = file_path_elenco.replace('.xlsx', ' Preenchido.xlsx')
                wb.save(output_path)
                # No Colab, precisamos disponibilizar o arquivo para download
                files.download(output_path)
            except Exception as e:
                print(f"Erro ao salvar o arquivo {file_path_elenco}: {str(e)}")
                return

    # Preencher a base de dados
    base = pd.ExcelFile(file_path_base)
    sheet_names = base.sheet_names
    base = pd.read_excel(file_path_base, sheet_name=sheet_names)

    for sheet in sheet_names[1:]:
        for d in range(len(base[sheet])):
            disciplina = base[sheet].loc[d, 'Disciplina (código)']

            if disciplina in solucao['Disciplina'].tolist():
                base[sheet].loc[d, 'Sala'] = None
                solucao_filtrada = solucao[solucao['Disciplina'] == disciplina]

                novo_valor = ['0', '0', '0', '0']
                for i in range(4):
                    for a in solucao_filtrada.index:
                        if base[sheet].loc[d, 'Horário ' + str(i+1)] == solucao_filtrada.loc[a, 'Horário']:
                            novo_valor[i] = str(solucao_filtrada.loc[a, 'Sala'])

                salas_fixadas = novo_valor[0]
                for s in novo_valor[1:]:
                    salas_fixadas += ', ' + str(s)

                base[sheet].loc[d, 'Sala'] = salas_fixadas

    try:
        output_path = file_path_base.replace('.xlsx', ' Preenchido.xlsx')
        with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
            for sheet in sheet_names:
                base[sheet].to_excel(writer, sheet_name=sheet, index=False)
        files.download(output_path)
    except Exception as e:
        print(f"Erro ao salvar o arquivo base: {str(e)}")
        return

    # Mensagem de sucesso
    if preencher_elenco:
        print("Sucesso! Os seguintes arquivos foram criados utilizando os Dados da Solução do Modelo:")
        for elenco in lista_elenco:
            print(f"- {os.path.basename(elenco.replace('.xlsx', ' Preenchido.xlsx'))}")
        print(f"- {os.path.basename(file_path_base.replace('.xlsx', ' Preenchido.xlsx'))}")
    else:
        print("Sucesso! O seguinte arquivo foi criado utilizando os Dados da Solução do Modelo:")
        print(f"- {os.path.basename(file_path_sol)}")
        print(f"- {os.path.basename(file_path_base.replace('.xlsx', ' Preenchido.xlsx'))}")
```
